## Investigating values using histogram and kde plots:

```python
def plot_data(df_column, loc):
    plt.subplot(1,3,loc)
    sns.histplot(df_column, kde=True)#,bins=20)

columns = ['National Income','Growth', 'Per person']
sns.set(rc={'figure.figsize': [15,5]}, font_scale=0.9, style='whitegrid')
for i,j in zip(range(3),columns):
    plot_data(df[j],i+1)
plt.show()
```

```python
def plot_data(df_column, loc):
    plt.subplot(1,3,loc)
    sns.histplot(data=df,x=df_column, hue='Continent', multiple='dodge')#,bins=20)

columns = ['National Income','Growth', 'Per person']
sns.set(rc={'figure.figsize': [15,5]}, font_scale=0.9, style='whitegrid')
for i,j in zip(range(3),columns):
    plot_data(j,i+1)
plt.show()
```

```python
def plot_data(df_column, loc):
    plt.subplot(1,3,loc)
    sns.histplot(data=df,x=df_column, hue='Continent')

columns = ['National Income','Growth', 'Per person']
sns.set(rc={'figure.figsize': [15,5]}, font_scale=0.9, style='whitegrid')
for i,j in zip(range(3),columns):
    plot_data(j,i+1)
plt.show()
```

*from this we can conclude:*

**1- For National Income:**
- The histogram is skewed left
- more than 30 countries (more than 50% of the countries) have income ranging through 50-500 X USDs (can also be concluded from the 2nd quartile)
- Asian and African countries dominate in lowest national incomes except for 1 or 2 countries in asia while the highest incomes are for North-american countries

**(this is very acceptable because most countries are consuming countries (Third world countries are most concentrated in afria and asia) while only a few countries are considered to be sovereign states of high economical power)**

**2-For Growth Rate:**
- This is a perfect bell-shaped histogam where most countries develop to have a growth rate ranging 2 or 3 times
